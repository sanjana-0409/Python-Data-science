# Chi Square Test

```python
import pandas as pd
import numpy as np

df = pd.read_csv("mtcars.csv")
```

```python
df.head()
```

```python
crosstable = pd.crosstab(df['cyl'],df['gear'])
```

```python
crosstable
```

```python
from scipy import stats
stats.chi2_contingency(crosstable)
```

```python
from scipy.stats import chi2_contingency
chi, p, dof, expected = chi2_contingency(crosstable)
```

```python
print(p)
```

```python
import pandas as pd
from scipy import stats
```

```python
#H0: mu1=mu2=m3, i.e., there is no significant difference between three diet plans
#H1: not all means are equal, i.e., there is a significant difference between three diet plans
#test= F-test(ANOVA test)
alpha=0.05
```

```python
Atkins=[6,2,3,4,2,3,3,2,7,8,10,4]
GM=[4,4,5,7,8,5,3,7,10,4,5,2]
SB=[6,7,5,6,8,7,8,9,6,5,5,12]
```

```python
import pandas as pd
data=pd.DataFrame({'Atkins':Atkins,'GM':GM,'SB':SB})
```

```python
from scipy import stats
stats.f_oneway(data['Atkins'],data['GM'],data['SB'])
```

```python
p=stats.f_oneway(data['Atkins'],data['GM'],data['SB'])[1]
if p<0.05:
  print('Reject Ho, i.e., there is a significant difference between three diet plans')
else:
  print('Accept Ho,i.e., there is no significant difference between three diet plans')
```
